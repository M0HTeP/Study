```python
class Set_x:
    __size = None
    data = None
    CONST_HASH = 262822338
    CONST_NEXT = 813871
    def __init__(self, size = 2 ** 22):
        """
        size: не задает реальный размер, но дает представление о ПОРЯДКЕ величины размера
        """
        self.__size = size
        self.data = [None] * self.__size
        
    def insert(self, x):
        """
        Ваш код тут:
            1) Взять индекс в который кладем
            2) Проверить пусто ли там или занято, возможно там заглушка после удаления
            3) Положить в свободное место элемент
            4) Увеличить размер множества по необходимости
            5) Уменьшить число свободных ячеек в счетчике
        """
        index = self.h(x)
        if self.data[index] == None:
          self.data[index] = x
        else:
          while True:
            if (self.data[index] == None or self.data[index] == 'x'):
              self.data[index] = x
              break
            index = self.next(index)
        self.__size -= 1
        if (len(self.data) >= self.__size * 2):
            self.__inc_capacity()
        
          

  
    def delete(self, x):
        """
        1) Берем начальный индекс
        2) Путешествуем по индексам до тех пор пока не найдем нужный элемент или не наткнемся на пустоту
        3) На место элемента x ставим заглушку
        """
        index = self.h(x)
        if self.data[index] == x:
          self.data[index] = 'x'

        else:
          while (self.data[index] != None):
            index = self.next(index)
            if index == x:
              self.data[index] = 'x'
              break


        
    def contains(self, x):
        """
        1) Берем начальный индекс
        2) Путешествуем по индексам до тех пор пока не найдем нужный элемент или не наткнемся на пустоту
        3) Если нашли то возвращаем True, если наткнулись на пустоту, то возвращаем false
        """
        index = self.h(x)
        if self.data[index] == x:
          return True
        while (self.data[index] != None):
          index = self.next(index)
          if index == x:
            return True
        return False



            
    def h(self, x):
        return (x + self.CONST_HASH) % len(self.data)
    
    def next(self, i):
        return (self.CONST_NEXT * i + 1) % len(self.data)
    
    def __inc_capacity(self):
        """
        1) Увеличить __size в 2 раза от текущего размера хранилища
        2) Положить старые данные в переменную
        3) Заполнить хранилище None 
        4) Пройтись по всем элементам старых данных и переложить их в новое хранилище в случае, если не None
        """
        n_data = self.data
        self.data = [None] * len(n_data) * 2
        self.__size = len(self.data)
        for i in n_data:
          if i != None:
            self.insert(i)

def main():
    s = Set_x()
    for i in range(int(input("Введите количество операций: "))):
        line = input(">>")
        cmd, d = line.split(' ')
        d = int(d)
        if cmd == 'insert':
            s.insert(d)
        elif cmd == 'exists':
            if s.contains(d):
                print('true\n')
            else:
                print('false\n')
        else:
            s.delete(d)
main()
```
